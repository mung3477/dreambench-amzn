### Code Backlogs

```python
unique_keys = set()

for item in df['images'].dropna():
    # If the item is a single dictionary
    if isinstance(item, dict):
        unique_keys.update(item.keys())
    # If the item is a list/array of dictionaries
    elif isinstance(item, (list, tuple)):
        for sub_item in item:
            if isinstance(sub_item, dict):
                unique_keys.update(sub_item.keys())

print(unique_keys)
```

```python
def extract_images(item, img_key = 'large'):
    if isinstance(item, dict):
        return item.get(img_key)
    return None

# Create a new column containing only the hi_res values
images = df['images'].apply(extract_images).apply(
    lambda lst: [item for item in lst if item is not None] if isinstance(lst, list) else lst
)
```

```python
image_num = images.apply(lambda l: len(l))
image_num.value_counts()
```
